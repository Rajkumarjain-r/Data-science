Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
url_for() function.

 function is used for URL building. It generates a URL for a given endpoint (view function) based on its name.

```python
from flask import Flask,url_for
app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "<H1>Welcome to ABC Corporation<H1>"

@app.route("/")
def show_details():
    data = "Company Name: ABC Corporation <br>Location: India <br>Contact Detail: 999-999-9999"
    return data

if __name__ == "__main__":
    with app.test_request_context():
        print(url_for('welcome'))  # Output: /welcome
```

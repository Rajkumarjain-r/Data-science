Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

Estimation in statistics refers to the process of using sample data to make inferences or predictions about the population parameters. Estimation statistics provides a way to estimate unknown population parameters based on information gathered from a sample.

Point Estimate:
A point estimate is a single value that serves as the best estimate of an unknown population parameter. It is obtained by using a sample statistic to estimate the corresponding population parameter. For example, the sample mean (x̄) is often used as a point estimate of the population mean (μ), and the sample proportion (p) is used as a point estimate of the population proportion (π).
Point estimates are straightforward and easy to compute. However, they have a limitation in that they do not provide information about the precision or uncertainty of the estimate. To address this limitation, interval estimates are used.

Interval Estimate:
An interval estimate provides a range of values within which the true population parameter is likely to lie. It takes into account the variability in the sample data and provides a measure of uncertainty associated with the estimate. Interval estimates are commonly expressed as confidence intervals.
A confidence interval consists of two values, the lower bound and the upper bound, and it is constructed around the point estimate. The level of confidence (e.g., 95% or 99%) indicates the probability that the interval will contain the true population parameter in repeated sampling.

```python
Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.
```

```python
import scipy.stats as stats

def hypothesis_test(sample1, sample2, alpha=0.05):
    # Perform the independent t-test
    t_statistic, p_value = stats.ttest_ind(sample1, sample2)

    # Check if the p-value is less than the significance level (alpha)
    if p_value < alpha:
        print("Reject the null hypothesis.")
        print("There is a significant difference between the two population means.")
    else:
        print("Fail to reject the null hypothesis.")
        print("There is no significant difference between the two population means.")

    print(f"T-Statistic: {t_statistic:.4f}")
    print(f"P-Value: {p_value:.4f}")

# Example usage:
# Assuming you have two sample datasets: sample1 and sample2
sample1 = [83, 76, 85, 88, 80, 92, 78, 81, 79, 85]
sample2 = [75, 78, 80, 72, 73, 77, 80, 75, 76, 74]

hypothesis_test(sample1, sample2)
```

```python
Q9. Generate a Student's t-distribution plot using Python's matplotlib library, with the degrees of freedom
parameter set to 10
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

degrees_of_freedom = 10


x = np.linspace(-5, 5, 500)  
t_distribution = stats.t.pdf(x, degrees_of_freedom)

# Plot the t-distribution
plt.plot(x, t_distribution, label=f"Degrees of Freedom = {degrees_of_freedom}")
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Student\'s t-Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

```python
Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two
random samples of equal size and a null hypothesis that the population means are equal.

```

```python
import numpy as np
from scipy.stats import ttest_ind

def two_sample_ttest(sample1, sample2):
    # Perform two-sample t-test
    t_stat, p_value = ttest_ind(sample1, sample2)

    # Print the results
    print("T-Statistic:", t_stat)
    print("P-Value:", p_value)

    # Interpret the results
    alpha = 0.05  # Significance level
    if p_value < alpha:
        print("Reject the null hypothesis: Population means are NOT equal.")
    else:
        print("Fail to reject the null hypothesis: Population means are equal.")

if __name__ == "__main__":
    # Replace these samples with your own data
    sample1 = np.array([1.5, 2.3, 3.1, 4.0, 5.2])
    sample2 = np.array([2.0, 2.8, 3.5, 4.5, 5.7])

    # Check if the samples have equal sizes
    if len(sample1) != len(sample2):
        print("Error: The two samples should have equal sizes.")
    else:
        two_sample_ttest(sample1, sample2)
```

Q12: What is t-statistic? State the formula for t-statistic.

The t-statistic is a measure used in hypothesis testing to quantify how many standard errors the sample mean differs from the hypothesized population mean. It is the fundamental component of the t-test, which assesses whether the difference between sample means or the sample mean itself is statistically significant.

The formula for calculating the t-statistic for comparing two independent sample means (two-sample t-test) is given by:

t = (x̄1 - x̄2) / √[(s1^2 / n1) + (s2^2 / n2)]

Where:

x̄1 is the sample mean of the first group (sample 1).
x̄2 is the sample mean of the second group (sample 2).
s1^2 is the sample variance of the first group (sample 1).
s2^2 is the sample variance of the second group (sample 2).
n1 is the sample size of the first group (sample 1).
n2 is the sample size of the second group (sample 2).

```python
Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a
standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight
is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5
pounds with a significance level of 0.01.
```

```python
import numpy as np
from scipy.stats import t

def one_sample_ttest(sample_mean, population_mean, sample_std, sample_size):
    # Calculate the t-statistic
    t_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

    # Degrees of freedom
    df = sample_size - 1

    # Calculate the critical t-value for a significance level of 0.01 and one-tailed test
    critical_t = t.ppf(0.01, df)

    # Calculate the p-value (one-tailed test)
    p_value = t.cdf(t_stat, df)

    return t_stat, p_value, critical_t

if __name__ == "__main__":
    # Given data
    sample_mean = 4.8
    population_mean = 5
    sample_std = 0.5
    sample_size = 25

    # Perform the one-sample t-test
    t_stat, p_value, critical_t = one_sample_ttest(sample_mean, population_mean, sample_std, sample_size)

    # Print the results
    print("T-Statistic:", t_stat)
    print("P-Value:", p_value)

    alpha = 0.01  # Significance level

    if t_stat < critical_t or p_value < alpha:
        print("Reject the null hypothesis: The true mean weight of the products is less than 5 pounds.")
    else:
        print("Fail to reject the null hypothesis: The true mean weight of the products is 5 pounds or more.")
```

```python
Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 =
30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean
score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two
groups are equal with a significance level of 0.01.
```

```python
import numpy as np
from scipy.stats import t

def two_sample_ttest(sample1_mean, sample2_mean, sample1_std, sample2_std, n1, n2):
    # Calculate the pooled standard deviation
    pooled_std = np.sqrt(((n1 - 1) * sample1_std**2 + (n2 - 1) * sample2_std**2) / (n1 + n2 - 2))

    # Calculate the t-statistic
    t_stat = (sample1_mean - sample2_mean) / (pooled_std * np.sqrt((1/n1) + (1/n2)))

    # Degrees of freedom
    df = n1 + n2 - 2

    # Calculate the critical t-value for a significance level of 0.01 and two-tailed test
    critical_t = t.ppf(0.995, df)

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - t.cdf(np.abs(t_stat), df))

    return t_stat, p_value, critical_t

if __name__ == "__main__":
    # Given data for the first group
    sample1_mean = 80
    sample1_std = 10
    n1 = 30

    # Given data for the second group
    sample2_mean = 75
    sample2_std = 8
    n2 = 40

    # Perform the two-sample t-test
    t_stat, p_value, critical_t = two_sample_ttest(sample1_mean, sample2_mean, sample1_std, sample2_std, n1, n2)

    # Print the results
    print("T-Statistic:", t_stat)
    print("P-Value:", p_value)

    alpha = 0.01  # Significance level

    if np.abs(t_stat) > critical_t or p_value < alpha:
        print("Reject the null hypothesis: The population means for the two groups are NOT equal.")
    else:
        print("Fail to reject the null hypothesis: The population means for the two groups are equal.")
```

```python
Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV
program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard
deviation of 1.5. Estimate the population mean with a 99% confidence interval.
```

```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample_mean, sample_std, sample_size, confidence_level):
    # Degrees of freedom
    df = sample_size - 1

    # Calculate the critical t-value for the given confidence level and two-tailed test
    critical_t = t.ppf(1 - (1 - confidence_level) / 2, df)

    # Calculate the margin of error
    margin_of_error = critical_t * (sample_std / np.sqrt(sample_size))

    # Calculate the lower and upper bounds of the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

if __name__ == "__main__":
    # Given data
    sample_mean = 4
    sample_std = 1.5
    sample_size = 50
    confidence_level = 0.99

    # Calculate the confidence interval
    lower_bound, upper_bound = confidence_interval(sample_mean, sample_std, sample_size, confidence_level)

    # Print the results
    print(f"Population Mean (99% Confidence Interval): {sample_mean:.2f} ± {upper_bound - sample_mean:.2f}")
    print(f"Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]")
```

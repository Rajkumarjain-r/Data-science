```python
Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
```

creates a figure and a grid of subplots with a single call, while providing reasonable control over how the individual plots are created. For more advanced use cases you can use GridSpec for a more general subplot layout 

```python
fig = plt.figure()
gs = fig.add_gridspec(4, hspace=.5)
axs = gs.subplots(sharex=False, sharey=False)
#fig.suptitle('Sharing both axes')
axs[0].plot(np.array([0, 1, 2, 3, 4, 5]), np.array([0, 100, 200, 300, 400, 500]))
axs[1].plot(np.array([0, 1, 2, 3, 4, 5]),np.array([50, 20, 40, 20, 60, 70]))
axs[2].plot(np.array([0, 1, 2, 3, 4, 5]), np.array([10, 20, 30, 40, 50, 60]))
axs[3].plot(np.array([0, 1,2, 3, 4, 5]), np.array([200, 350, 250, 550, 450, 150]))
```

```python
Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
box plot is used to find the mean, 25 percentile, 50 percentile, 75 percentile, median, outliers
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
columns = [box1,box2]
```

```python
fig, ax = plt.subplots()
ax.boxplot(columns)
plt.show()
```

```python
Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

```

```python
import pandas as pd
import numpy as np
data_size = 50

# Generate random data for 'Study Time' and 'Exam Score'
np.random.seed(42)  # For reproducibility
study_time_data = np.random.randint(1, 10, data_size)  # Random study time between 1 and 10 hours
exam_score_data = np.random.randint(40, 100, data_size)  # Random exam score between 40 and 100

# Create the DataFrame
df = pd.DataFrame({
    'Study Time': study_time_data,
    'Exam Score': exam_score_data
})
```

```python
df
```

```python
df.corr(method='pearson')
```

```python
the between study time and excamsure the pearson correlation is +ve meaning study time proportional to exam score
```

```python
Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.
```

```python
np.random.seed(42)  
sleep_data = np.random.randint(4, 10, data_size)  # Random sleep hours between 4 and 10
job_satisfaction_data = np.random.randint(1, 11, data_size)  # Random job satisfaction levels between 1 and 10
df = pd.DataFrame({
    'Sleep': sleep_data,
    'Job Satisfaction': job_satisfaction_data
})
```

```python
df.corr(method='spearman')
```

```python
between sleep and job satisfication, they are positvely corelated
```

```python
data_size = 50

# Generate random data for 'Exercise Hours' and 'BMI'
np.random.seed(42)  # For reproducibility
exercise_hours_data = np.random.randint(1, 10, data_size)  # Random exercise hours between 1 and 10
bmi_data = np.random.uniform(18.5, 35.0, data_size)  # Random BMI between 18.5 and 35.0

# Create the DataFrame
df = pd.DataFrame({
    'Exercise Hours': exercise_hours_data,
    'BMI': bmi_data
})
```

```python
df.corr(method= 'spearman')
```

```python
df.corr()
```

```python
both give same results that is BMI and Execise hours are negatively corelated
```

```python
Q4. A researcher is interested in examining the relationship between the number of hours individuals
spend watching television per day and their level of physical activity. The researcher collected data on
both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between
these two variables.
```

```python
# Sample data size
data_size = 50

# Generate random data for 'TV Hours' and 'Physical Activity Level'
np.random.seed(42)  # For reproducibility
tv_hours_data = np.random.uniform(0, 5, data_size)  # Random TV hours between 0 and 5
physical_activity_data = np.random.uniform(1, 10, data_size)  # Random physical activity level between 1 and 10

# Create the DataFrame
df = pd.DataFrame({
    'TV Hours': tv_hours_data,
    'Physical Activity': physical_activity_data
})
```

```python
df.head()
```

```python
# Calculate the Pearson correlation coefficient
pearson_corr = df['TV Hours'].corr(df['Physical Activity'])

print("\nPearson Correlation Coefficient:", pearson_corr)
```

```python
Q6. A company is interested in examining the relationship between the number of sales calls made per day
and the number of sales made per week. The company collected data on both variables from a sample of
30 sales representatives. Calculate the Pearson correlation coefficient between these two variables.
```

```python
# Sample data size
data_size = 30

# Generate random data for 'Sales Calls per Day' and 'Sales per Week'
np.random.seed(42)  # For reproducibility
sales_calls_per_day = np.random.randint(20, 50, data_size)  # Random sales calls between 20 and 50
sales_per_week = np.random.randint(5, 20, data_size)  # Random sales per week between 5 and 20

# Create the DataFrame
df = pd.DataFrame({
    'Sales Calls per Day': sales_calls_per_day,
    'Sales per Week': sales_per_week
})
```

```python
df.corr(method = 'pearson')
```

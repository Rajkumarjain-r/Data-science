```python
Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.
```

```python
import math

def normal_pdf(x, mean, std_dev):
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = -0.5 * ((x - mean) / std_dev) ** 2
    pdf_value = coefficient * math.exp(exponent)
    return pdf_value

mean = 5.0
std_dev = 2.0
x_value = 7.0

pdf_result = normal_pdf(x_value, mean, std_dev)
print(f"PDF at x={x_value}: {pdf_result}")
```

```python
Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.
```

```python
from scipy.stats import poisson

def poisson_cdf(x, mean):
    cdf_value = poisson.cdf(x, mean)
    return cdf_value

mean = 3.5
x_value = 5

cdf_result = poisson_cdf(x_value, mean)
print(f"CDF at x={x_value}: {cdf_result}")
```

Nature of Events:

Binomial Distribution: The Binomial distribution is used when each trial has two possible outcomes, typically labeled as "success" and "failure." The probability of success remains constant across all trials, and the trials are independent.
Poisson Distribution: The Poisson distribution is used to model the number of events that occur in a fixed interval of time or space. It is suited for counting the occurrences of rare events in continuous time or space.
    

Probability of Success:

Binomial Distribution: The probability of success in each trial remains constant and is denoted as "p."
Poisson Distribution: The Poisson distribution does not directly use a probability of success. Instead, it is characterized by the average rate of occurrence (λ, lambda) of events in the fixed interval of time or space.

Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

Binomial Distribution:
In the Binomial distribution, the mean (μ) and variance (σ^2) are related as follows:
Mean (μ) = n * p

Variance (σ^2) = n * p * (1 - p)

where:

n is the number of trials or sample size.
p is the probability of success in each trial.
In the Binomial distribution, the mean is equal to the product of the number of trials and the probability of success, and the variance is equal to the product of the number of trials, the probability of success, and the probability of failure (1 - p).

The relationship between mean and variance in the Binomial distribution is intuitive. The more trials there are (higher "n") or the higher the probability of success (higher "p"), the larger both the mean and variance will be. When the probability of success is low, the variance becomes larger compared to the mean, indicating more variability in the outcomes.

Poisson Distribution:
In the Poisson distribution, the mean (λ, lambda) and variance (σ^2) are related as follows:
Mean (λ) = Variance (σ^2) = λ

where:

λ (lambda) is the average rate of occurrence of events in the fixed interval.
